```python
import json

fd = open("records.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)



print("Welcome to shop")


ui_prod = input("Enter the product Id: ")
ui_quan = int(input("Enter the quantity of the Product: "))

for i in records.keys():
    while ui_prod == i:
        if ui_quan <= records[i]['quan']:
            print("****************************************")
            print("             product:",records[ui_prod]['name'])
            print("               price:",records[ui_prod]['price'])
            print("         Expiry date:",records[ui_prod]['expiry'])
            print(" Rating:",records[ui_prod]['rating'])
            print("****************************************")
            print("----------BILLING DETAILS:---------- ")
            print("Total Amount: ",records[ui_prod]['price']*ui_quan) 
            print("Enjoy Your Day")
            print("Thank You. Visit Again")
            print("**********************************")
            records[ui_prod]["quan"] = records[ui_prod]["quan"] - ui_quan
        else:
            print("Sorry. Stock of product you enterd is: ",records[ui_prod]['quan'],"\nPlease accept our apologies for this inconvenience")
        break
if ui_prod not in records.keys():
   print("Sorry. No product found for ID: ",ui_prod,"\Please enter valid product ID")
    
    
js = json.dumps(records)

fd = open("records.json",'w')
fd.write(js)
fd.close()
```
